```python
import matplotlib.pyplot as plt


#define some data
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]


#plot data
plt.plot(x, y, linestyle="dashdot", marker="*", color="blue", label = 'Line Example')	

#configure  X axes
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])

#configure  Y axes
plt.ylim(19.8,21.2)
plt.yticks([20, 21, 20.5, 20.8])

#labels
plt.xlabel("this is X")
plt.ylabel("this is Y")

# add legend
plt.legend()

#title
plt.title("Simple plot")

#show plot
plt.show()
```

```python
import matplotlib.pyplot as plt


#define some data
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

x1 = [2,4,6,8]
y1 = [12, 25, 18, 30.8]


#plot data
plt.subplot(2, 2, 1)
plt.plot(x, y, linestyle="dotted", marker='^', color="red", label = 'Red Line')	
plt.subplot(2, 2, 2)
plt.plot(x1, y1, linestyle="dashdot", marker="*", color="blue", label = 'Blue Example')

#configure  X axes


# add legend
plt.legend()

#title
plt.title("Simple plot")

#show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame.from_csv('product.csv', index_col = False)


#plot data
plt.plot(data['Year'], data['Product 1'], linestyle="dotted", marker='^', color="red", label = 'Product 1')

plt.plot(data['Year'], data['Product 2'], linestyle="solid", marker='o', color="green", label = 'Product 2')

plt.plot(data['Year'], data['Product 3'], linestyle="dashed", marker='<', color="blue", label = 'Product 3')



#labels
plt.xlabel("Year")
plt.ylabel("Number in units")

# add legend
plt.legend()

#title
plt.title("Poduct Analysis")

#show plot
plt.show()
```
