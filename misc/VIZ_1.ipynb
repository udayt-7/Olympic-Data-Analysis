```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

#dx = pd.read_csv("dataset.csv")
dm = pd.read_csv('athlete_events.csv')
df = pd.read_csv('rd1.csv')
#print(df)

ds = pd.read_csv('summer.csv')
#print(ds)

dw = pd.read_csv('winter.csv')
#print(dw)
dm
```

```python
x = dm.groupby('Year')['ID'].count()
#x=t1.groupby("Year")["ID"].count()
population=pd.DataFrame({'Year':x.index,'player_count':x.values})
objects = population['Year']
y_pos = np.arange(len(objects))
performance=population['player_count']
plt.figure(figsize=(15, 4))
plt.bar(y_pos, performance, align = 'center',alpha= 1)
plt.xticks(y_pos, objects,rotation=90)
plt.ylabel('Participants')
plt.xlabel('year')
plt.title('Partcipants by year')
plt.show()
```

```python

sport=df.groupby("Sport")["Medal"].count().nlargest(5)
s =pd.DataFrame({'Sport':sport.index,'Medal':sport.values})
top5 = s.sort_values('Medal', ascending=False).head(5)
top5

sport = s['Sport']
medal_data = s["Medal"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(medal_data, labels=sport, explode=explode, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Top 5 Sports")
plt.show()

```

```python


medal=df.groupby("NOC")["Medal"].count().nlargest(5)
M=pd.DataFrame({'NOC':medal.index,'Medal':medal.values})
# objects = M['NOC']
# y_pos = np.arange(len(objects))
# performance=M['Medal']
# plt.figure(figsize=(13,4))
# plt.bar(y_pos, performance, align = 'center',alpha=0.65)
# plt.xticks(y_pos, objects,rotation=90)
# plt.ylabel('Medal_Count')
# plt.xlabel('Sport')
# plt.title('Top 10 country with most medals')
# plt.show()

country_data = M['NOC']
medal_data = M["Medal"]
colors =  ['gold','brown','cyan','blue', 'red']
explode = (0, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors = colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Top 5 countries overall")
plt.show()

# slices_countries = m[]
# activities = ['Sleep', 'Work']
# colors = ['r', 'g']

# plt.pie(slices_hours, labels=activities, colors=colors, startangle=90, autopct='%.1f%%')
# plt.show()
```

```python
yearsplit = df[df['Year'].between(1896,1930, inclusive = True)].sort_values(['Year','Season'],ascending = False)
medal=yearsplit.groupby("NOC")["Medal"].count().nlargest(5)
t1=pd.DataFrame({'NOC':medal.index,'Medal':medal.values})

country_data = t1['NOC']
medal_data = t1["Medal"]
colors =  ['gold','blue','red','brown', 'teal']
explode = (0, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors = colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Top 5 in first 40 years [1896 - 1930]")
plt.show()
```

```python
yearsplit = df[df['Year'].between(1930,1970, inclusive = True)].sort_values(['Year','Season'],ascending = False)
medal=yearsplit.groupby("NOC")["Medal"].count().nlargest(5)
t1=pd.DataFrame({'NOC':medal.index,'Medal':medal.values})

country_data = t1['NOC']
medal_data = t1["Medal"]
colors =  ['gold','brown','teal','red', 'lightgreen']
explode = (0, 0, 0, 0, 0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors = colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Top 5 from [1931 - 1971]")
plt.show()
```
